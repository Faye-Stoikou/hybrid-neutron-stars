```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.cm as cm

#---------------------------------------------------------

#Constants 
Msun_in_km = 1.474  #In units where G=c=1 (Geometrized)
MeVfm3_to_Jm3 = 1.6022e32 
Jm3_to_1km2 = 8.234568e-39 
MeVfm3_to_1km2 = MeVfm3_to_Jm3 * Jm3_to_1km2

#---------------------------------------------------------

#Equations of State APR-1 and the Crust Equations  
def MDI_1(P):
    return 4.1844*(P**0.81449) + 95.00135*(P**0.31736)

def APR_1(P):
    return 0.000719964*(P**1.85898) + 108.975*(P**0.340074)

def Crust_1(P):
    return 0.00873 + 103.17338*(1-np.exp(-P/0.38527)) + 7.34979*(1-np.exp(-P/0.01211))

def Crust_2(P):
    return  0.00015 + 0.00203*(1-np.exp(-P*344827.5)) + 0.10851*(1-np.exp(-P*7692.3076))
    
def Crust_3(P):                                                             
    return 0.0000051*(1-np.exp(-P*0.2373e10)) + 0.00014*(1-np.exp(-P*0.4020e8))

#---------------------------------------------------------

#Energy Desnity of Pure Hadronic Matter 
def den_en_hadronic(P):
    if P < 4.1725e-08:
        return Crust_3(P)
    elif P < 9.34375e-05:
        return Crust_2(P)
    elif P < 0.184:
        return Crust_1(P)
    else:
        return APR_1(P)

#Energy Density of Hybrid Matter 
def den_en_hybrid(P, Ptr, DE, cs2=1, check_seidov=False):
    if P < 4.1725e-08:
        return Crust_3(P)
    elif P < 9.34375e-05:
        return Crust_2(P)
    elif P < 0.184:
        return Crust_1(P)
    elif P < Ptr:
        return APR_1(P)
    else:
        en_tr = APR_1(Ptr)
        if check_seidov:
            DE_crit = 0.5*en_tr + 1.5*Ptr
            if DE > DE_crit:
                print("Seidov satisfied → twin branch")
            else:
                print("Seidov unsatisfied → no twin branch")
        return en_tr + DE + (P - Ptr)/cs2  #cs^2 = 1 (maximally stiff)
            
```
